```python
usd_krw_data = usd_krw_data.rename(columns=usd_krw_data.iloc[0]).drop(usd_krw_data.index[0])
```

```python
usd_krw_data = usd_krw_data.iloc[:,:10]
```

```python
usd_krw_data
```

```python
usd_krw_data = usd_krw_data[['날짜','실업자','실업률(%)','청년실업자','청년실업률(%)']]
```

```python
usd_krw_data.reset_index(drop=True,inplace=True)
```
